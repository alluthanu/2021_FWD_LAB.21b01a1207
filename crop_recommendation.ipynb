```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Crop_recommendation.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df['label'].value_counts()
```

```python
x=df.drop('label',axis=1)
```

```python
y=df['label']
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
x_train.info()
```

```python
x_test.info()
```

```python
y_train.info()
```

```python
y_test.info()
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
logistic_reg=accuracy_score(y_test,y_pred)
```

```python
print('logistic accuracy is '+str(logistic_reg))
```

```python
from sklearn.tree import DecisionTreeClassifier
model_Dectree=DecisionTreeClassifier()
model_Dectree.fit(x_train,y_train)
```

```python
y_pred3=model_Dectree.predict(x_test)
```

```python
Decison_acc=accuracy_score(y_test,y_pred3)
print(Decison_acc)
```

```python
from sklearn.ensemble import RandomForestClassifier
model_random=RandomForestClassifier()
model_random.fit(x_train,y_train)
```

```python
y_pred4=model_random.predict(x_test)
```

```python
random_acc=accuracy_score(y_test,y_pred4)
print(random_acc)
```

```python
import joblib
filename='crop_app'
joblib.dump(model_random,'crop_app')
```
